```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
type(iris)
```

```python
print(list((iris.feature_names)))
```

```python
print(iris.target)
```

```python
iris.data
```

```python
df=pd.DataFrame(iris.data)
```

```python
df.columns=iris.feature_names
```

```python
df['label']=iris.target
```

```python
df.head(10)
```

```python
df.count()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['sepal length (cm)'],df['sepal width (cm)'])
plt.title('Sepal')
plt.xlabel('lenght')
plt.ylabel('width')
```

```python
plt.bar(df['petal length (cm)'],df['petal width (cm)'])
plt.title("petal")
plt.xlabel('length')
plt.ylabel('width')
```

```python
plt.hist(df['petal length (cm)'])
plt.hist(df['petal width (cm)'])
```
