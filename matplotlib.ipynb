```python
x1_point=np.array([0,1,2,3])
y1_point=np.array([3,8,1,10])
x2_point=np.array([0,1,2,3])
y2_point=np.array([6,2,7,11])
plt.plot(x1_point,y1_point)
plt.plot(x2_point,y2_point)
plt.title("Sample Graph")
plt.xlabel("num1")
plt.ylabel("num2")
plt.show()
```

```python
x1_point=np.array([0,1,2,3])
y1_point=np.array([3,8,1,10])
x2_point=np.array([0,1,2,3])
y2_point=np.array([6,2,7,11])
plt.plot(x1_point,y1_point)
plt.plot(x2_point,y2_point)

font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'red','size':10}
plt.title("Sample Graph",fontdict=font1)
plt.xlabel("Roll_No",fontdict=font2)
plt.ylabel("Marks",fontdict=font1)
plt.plot(x1_point,y1_point)
plt.plot(x2_point,y2_point)
#plt.grid()
plt.grid(color='green',linestyle='--',linewidth=0.5)
plt.show()
```
